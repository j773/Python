# Matplotlib

Es una librería para crear gráficos. Se puede usar tanto dentro de un sistema Python (a través de la API orientada a objetos) como también en forma más conveniente en los portátiles Jupyter e IPython a través de la interfaz Pyplot

matplotlib.pyplot es una colección de funciones de estilo de comando que hacen que matplotlib funcione como MATLAB. Cada función de pyplot hace algún cambio en una figura: por ejemplo, crea una figura, crea un área de trazado en una figura, traza algunas líneas en un área de trazado, decora el trazado con etiquetas, etc. En matplotlib.pyplot se conservan varios estados en la función llamadas, para que se mantenga al tanto de cosas como la figura actual y el área de trazado, y las funciones de trazado se dirigen a los ejes actuales (tenga en cuenta que los "ejes" aquí y en la mayoría de los lugares en la documentación se refieren a la parte de los ejes de una figura y no el término matemático estricto para más de un eje).

### plt.subplots()

Devuelve una tupla que contiene una figura y objeto(s) de ejes. Por lo general se usa con la siguiente instrucción: __fig,axes=plt.subplots()__. 
Si queremos guardar una figura, lo podemos hacer con: fig.savefig('nombredelagrafica.png')

```python
fig,axes=plt.subplots()
axes.plot(x,funcion1(x),"*y",label="Funcion f(x)")
fig.savefig("grafica1.png") #Para guardar la figura
```

### Trabajando con texto

El comando text() se puede usar para agregar texto en una ubicación arbitraria, y xlabel(), ylabel()y title() se usa para agregar texto en las ubicaciones indicadas.

```python
n = np.random.randn(100000)
print(np.min(n))
print(np.max(n))
fig, axes = plt.subplots(1, 2, figsize=(12,4))

axes[0].hist(n,50,density=True, facecolor='c', alpha=0.9)
axes[0].set_title("Default histogram")
axes[0].set_xlim((min(n), max(n)))

axes[1].hist(n, cumulative=True, bins=6)
axes[1].set_title("Cumulative detailed histogram")
axes[1].set_xlim((min(n), max(n)));
```

También se puede escribir expresiones en latex rodeando la expresión en latex con el signo de dólar

```python
fig, ax = plt.subplots()

ax.plot(x, x**2, label=r"$x = \alpha^2$")
ax.plot(x, x**3, label=r"$y = \alpha^3$")
ax.legend(loc=2) # upper left corner
ax.set_xlabel(r'$\alpha$', fontsize=18)
ax.set_ylabel(r'$y$', fontsize=18)
ax.set_title('title');
```

### Anotando texto

Los usos del comando text () anterior colocan el texto en una posición arbitraria en los ejes. Un uso común del texto es anotar algunas características de la gráfica, y el método annotate () proporciona una funcionalidad de ayuda para facilitar las anotaciones. En una anotación, hay dos puntos a considerar: la ubicación que se anota representada por el argumento xy y la ubicación del texto xytext. Ambos de estos argumentos son (x, y) tuplas.

```python
ax = plt.subplot(111)

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

plt.annotate('punto máximo', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='g', shrink=0.05),
            )

plt.ylim(-2,2)
plt.show()
```

### Ejes logarítmicos y otros ejes no lineales 

```python
from matplotlib.ticker import NullFormatter  # useful for `logit` scale

# Fixing random state for reproducibility
np.random.seed(19680801)

# make up some data in the interval ]0, 1[
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

# plot with various axes scales
plt.figure(1)

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)


# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)


# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthreshy=0.01)
plt.title('symlog')
plt.grid(False)

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)
# Format the minor tick labels of the y-axis into empty strings with
# `NullFormatter`, to avoid cumbering the axis with too many labels.
plt.gca().yaxis.set_minor_formatter(NullFormatter())
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)

plt.show()
```

### Diagramas de dispersión

Matplotlib tiene una función incorporada para crear diagramas de dispersión llamados scatter(). Un diagrama de dispersión es un tipo de gráfico que muestra los datos como una colección de puntos. La posición de un punto depende de su valor bidimensional, donde cada valor es una posición en la dimensión horizontal o vertical.

```python
x=np.random.randn(100)
y=np.random.randn(100)

s=200*np.random.randn(100)#Para el área de los circulos
c=np.random.randn(100)#Para el color de cos circulos

plt.scatter(x,y,s,c,cmap=plt.cm.Accent)
plt.title("Uso de scatter()")
```

Matplotlib puede crear parcelas 3d. Hacer un diagrama de dispersión en 3D es muy similar a crear un 2D, solo algunas pequeñas diferencias. En algunas ocasiones, un diagrama de dispersión 3d puede ser una mejor visualización de datos que un gráfico 2d. Para crear gráficos 3d, necesitamos importar axes3d.

```python
from mpl_toolkits.mplot3d import axes3d

# Creación de la data
N = 60
g1 = (0.6 + 0.6 * np.random.rand(N), np.random.rand(N),0.4+0.1*np.random.rand(N))
g2 = (0.4+0.3 * np.random.rand(N), 0.5*np.random.rand(N),0.1*np.random.rand(N))
g3 = (0.3*np.random.rand(N),0.3*np.random.rand(N),0.3*np.random.rand(N))

data = (g1, g2, g3)
colors = ("red", "green", "blue")
groups = ("cafe", "té", "agua") 

# Creacón de la gráfica
fig = plt.figure()
ax = fig.gca(projection='3d')

for data, color, group in zip(data, colors, groups):
    x, y, z = data
    ax.scatter(x, y, z, alpha=0.8, c=color, edgecolors='none', s=30, label=group)

plt.title('Matplot 3d scatter plot')
plt.legend(loc=2)
plt.show()
```

### Inserción de Gráficos

```python
x = np.linspace(0, 5, 10)
y = x ** 2

fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes

# main figure
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('title')

# insert
axes2.plot(y, x, 'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('insert title');
```

### Tamaño y resolución de una figura

Matplotlib permite que se especifique la resolución y el tamaño de la figura cuando se crea el objeto Figure, utilizando los argumentos de las palabras clave figsize y dpi. Figsize es una tupla del ancho y alto de la figura en pulgadas, y dpi es el de puntos por pulgada (pixel por pulgada). Para crear una figura de 800x400 píxeles, 100 puntos por pulgada, podemos hacer:

```python
fig = plt.figure(figsize=(8,4), dpi=100)

fig, axes = plt.subplots(figsize=(12,3))

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

### Otros ploteos en 2D

### Observación 2

Pyplot crea las figuaras y los ejes de forma implícita y automática.

Pylab combina la funcionalidad pyplot (para trazar) con la funcionalidad numpy (para matemáticas y para trabajar con matrices) en un solo espacio de nombres, haciendo que ese espacio de nombres (o entorno) sea aún más parecido a MATLAB. Por ejemplo, uno puede llamar a las funciones sin y cos como lo haría en MATLAB, además de tener todas las características de pyplot.

```python
from pylab import *
```

```python
x = np.linspace(0, 5, 10)
y = x ** 2
```

```python
figure()
plot(x, y, 'r')
xlabel('x')
ylabel('y')
title('title')
show()
```

La mayoría de las funciones relacionadas con el trazado en MATLAB están cubiertas por el módulo pylab. Por ejemplo, subplot y selección de color / símbolo:

```python
subplot(1,2,1)
plot(x, y, 'r--')
subplot(1,2,2)
plot(y, x, 'g*-');
```
